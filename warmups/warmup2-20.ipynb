```python
from matplotlib import pyplot as plt 
import math
import numpy as np

fig1, ax1 = plt.subplots(1, 1) 
fig2, ax2 = plt.subplots(1,1)

z = np.linspace(0.01,5)
t = np.linspace(-1,.999)


ax1.set(xlabel = r"$z$", 
       ylabel = r"$f(z)$")


ax2.set(xlabel = r"$z$", 
       ylabel = r"$g(z)$")

ax1.plot(z, -1*np.log(z))
ax2.plot(t, -1*np.log(1-t))
```
